```python
documents = [
    "Le compost améliore la fertilité des sols en apportant de la matière organique.",
    "Le paillage réduit l'évaporation de l'eau et protège le sol contre l'érosion.",
    "La rotation des cultures permet de réduire la pression des maladies et des ravageurs.",
    "L'association du niébé et du mil améliore la gestion de l'azote dans certaines régions sahéliennes.",
    "L'utilisation d'extraits naturels de neem permet de lutter contre certains ravageurs.",
    "Les pratiques agroécologiques favorisent la biodiversité et la résilience des exploitations agricoles."
]

for doc in documents:
    print(doc)
```

# les embeddings

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

embeddings = model.encode(documents)

print(embeddings.shape)
```

# recherche sémantique

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

question = "Comment améliorer la fertilité du sol ?"

question_embedding = model.encode([question])

scores = cosine_similarity(
    question_embedding,
    embeddings
)

best_idx = np.argmax(scores)

print("Question :", question)
print()
print("Document trouvé :")
print(documents[best_idx])
```

# mini assistant

```python
while True:

    question = input("Votre question : ")

    if question.lower() == "quit":
        break

    question_embedding = model.encode([question])

    scores = cosine_similarity(
        question_embedding,
        embeddings
    )

    best_idx = np.argmax(scores)

    print("\nRéponse :")
    print(documents[best_idx])
    print()
```
